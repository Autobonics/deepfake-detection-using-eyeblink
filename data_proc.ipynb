```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
user=pd.read_csv('user_data.csv')
user
```

```python
original=pd.read_csv('original_data.csv')
original=original.merge(user)
original=original[original.columns[[0,1,2,4,5,6,3]]]
original['label']=0
original
```

```python
fake=pd.read_csv('fake_data.csv')
fake=fake.merge(user)
fake=fake[fake.columns[[0,1,2,4,5,6,3]]]
fake['label']=1
fake
```

```python
res=pd.concat([original,fake]).reset_index(drop=True)
res.hist(bins=50,figsize=[20,20])
plt.show()
```

```python
res.to_csv('final_dataset.csv',index=False)
```
